```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
x = np.arange(0.01, 5.01, 0.01)
y = np.zeros(500)
d = 3
b = d/2-1
for i in range(100):
    for j in range(500):
        mu=x[j]
        if i==0:
            y[j] = -(i+d/2)*np.log(1-b/(i+1+b)/2)*np.exp(-mu)
        else:
            y[j] += (-(i+d/2)*np.log(1-b/(i+1+b)/2)-i*(np.log(1+b/i/2)))*np.exp(-mu)*mu**i/np.math.factorial(i)   
plt.plot(x, y, label="d=3",color='black')

y = np.zeros(500)
d = 5
b = d/2-1
for i in range(100):
    for j in range(500):
        mu=x[j]
        if i==0:
            y[j] = -(i+d/2)*np.log(1-b/(i+1+b)/2)*np.exp(-mu)
        else:
            y[j] += (-(i+d/2)*np.log(1-b/(i+1+b)/2)-i*(np.log(1+b/i/2)))*np.exp(-mu)*mu**i/np.math.factorial(i) 
plt.plot(x, y, label="d=5",color='black')

y = np.zeros(500)
d = 8
b = d/2-1
for i in range(100):
    for j in range(500):
        mu=x[j]
        if i==0:
            y[j] = -(i+d/2)*np.log(1-b/(i+1+b)/2)*np.exp(-mu)
        else:
            y[j] += (-(i+d/2)*np.log(1-b/(i+1+b)/2)-i*(np.log(1+b/i/2)))*np.exp(-mu)*mu**i/np.math.factorial(i)  
plt.plot(x, y, label="d=8",color='black')

y = np.zeros(500)
d = 12
b = d/2-1
for i in range(100):
    for j in range(500):
        mu=x[j]
        if i==0:
            y[j] = -(i+d/2)*np.log(1-b/(i+1+b)/2)*np.exp(-mu)
        else:
            y[j] += (-(i+d/2)*np.log(1-b/(i+1+b)/2)-i*(np.log(1+b/i/2)))*np.exp(-mu)*mu**i/np.math.factorial(i) 
plt.plot(x, y, label="d=12",color='black')
plt.xlabel(chr(956),fontsize=16)
plt.ylabel("Risk difference",fontsize=16)
plt.xlim(0, 5)
plt.ylim(0, 3.9)
plt.text(0.2,0.28,'d=3',fontsize=14)
plt.text(0.5,0.7,'d=5',fontsize=14)
plt.text(1,1.24,'d=8',fontsize=14)
plt.text(2,1.8,'d=12',fontsize=14)
plt.savefig('figure1.eps')
#plt.legend()
plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.special import gamma
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(10,6))
plt.subplots_adjust(left=None, bottom=None, right=None, top=None,wspace=0.3)
x = np.arange(0.01, 5.01, 0.01)
y = np.zeros(500)
d = 3
b = d/2-1
for i in range(150):
    for j in range(500):
        mu=x[j]
        if i==0:
            y[j] = -(i+d/2)*np.log(1-b/(i+1+b)/2)*np.exp(-mu)
        else:
            y[j] += (-(i+d/2)*np.log(1-b/(i+1+b)/2)-i*(np.log(1+b/i/2)))*np.exp(-mu)*mu**i/np.math.factorial(i)   
axes[0].plot(x, np.log(y), label="Empirical Bayes 1",color='blue')

y = np.zeros(500)
b = d-2
for i in range(150):
    for j in range(500):
        mu=x[j]
        if i==0:
            y[j] = -(i+d/2)*np.log(1-b/(i+1+b)/2)*np.exp(-mu)
        else:
            y[j] += (-(i+d/2)*np.log(1-b/(i+1+b)/2)-i*(np.log(1+b/i/2)))*np.exp(-mu)*mu**i/np.math.factorial(i)   
axes[0].plot(x, np.log(y), label="Empirical Bayes 2",color='blue',linestyle = 'dotted')

y = np.zeros(500)+(d/2-1)*np.log(2)
for i in range(150):
    for j in range(500):
        mu=x[j]
        y[j] += (np.log(gamma(i+d/2))-np.log(gamma(i+1)))*np.exp(-mu)*mu**i/np.math.factorial(i)
        y[j] -= (np.log(gamma(i+d/2))-np.log(gamma(i+1)))*np.exp(-2*mu)*(2*mu)**i/np.math.factorial(i)
axes[0].plot(x, np.log(y), label="Shrinkage prior",color='red',linestyle = 'dashed')

axes[0].set_xlim(0, 5)
#plt.ylim(0, 0.45)
axes[0].set_xlabel(chr(956),fontsize=16)
axes[0].set_ylabel("Log risk difference",fontsize=16)

y = np.zeros(500)
d = 8
b = d/2-1
for i in range(150):
    for j in range(500):
        mu=x[j]
        if i==0:
            y[j] = -(i+d/2)*np.log(1-b/(i+1+b)/2)*np.exp(-mu)
        else:
            y[j] += (-(i+d/2)*np.log(1-b/(i+1+b)/2)-i*(np.log(1+b/i/2)))*np.exp(-mu)*mu**i/np.math.factorial(i)   
axes[1].plot(x, np.log(y), label="Empirical Bayes 1",color='blue')

y = np.zeros(500)
b = d-2
for i in range(150):
    for j in range(500):
        mu=x[j]
        if i==0:
            y[j] = -(i+d/2)*np.log(1-b/(i+1+b)/2)*np.exp(-mu)
        else:
            y[j] += (-(i+d/2)*np.log(1-b/(i+1+b)/2)-i*(np.log(1+b/i/2)))*np.exp(-mu)*mu**i/np.math.factorial(i)   
axes[1].plot(x, np.log(y), label="Empirical Bayes 2",color='blue',linestyle = 'dotted')

y = np.zeros(500)+(d/2-1)*np.log(2)
for i in range(150):
    for j in range(500):
        mu=x[j]
        y[j] += (np.log(gamma(i+d/2))-np.log(gamma(i+1)))*np.exp(-mu)*mu**i/np.math.factorial(i)
        y[j] -= (np.log(gamma(i+d/2))-np.log(gamma(i+1)))*np.exp(-2*mu)*(2*mu)**i/np.math.factorial(i)
axes[1].plot(x, np.log(y), label="Shrinkage prior",color='red',linestyle = 'dashed')

axes[1].set_xlim(0, 5)
#plt.ylim(0, 0.45)
axes[1].set_xlabel(chr(956),fontsize=16)
axes[1].set_ylabel("Log risk difference",fontsize=16)
axes[0].set_title('(a)',loc='left',fontsize=16)
axes[1].set_title('(b)',loc='left',fontsize=16)

plt.legend()
plt.savefig('figure2.eps')
plt.show()
```

```python
y = np.zeros(20001)
for i in range(1,20001):
    lambd=0.001*i
    f=np.log(0.5/lambd)*np.exp(-lambd)*lambd
    a=0
    for x in range(1,max(100,round(lambd)*3)):
        a+=np.log(lambd/x)
        f+=np.log((x+0.5)/lambd)*np.exp(a-lambd)*lambd
    y[i]=f
```
